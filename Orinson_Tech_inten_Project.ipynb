```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
cols=["driveway","recroom",	"fullbase",	"gashw","airco","garagepl","prefarea"]
for f in cols:
    df[f]=le.fit_transform(df[f])
```

```python
df.head()
```

```python
df.corr()
```

```python
x=df[["lotsize"]]
y=df.price
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
le=LinearRegression()
le.fit(x_train,y_train)
```

```python
le.coef_
```

```python
le.intercept_
```

```python
ypred=le.predict(x_test)
```

```python
ypred
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
sr2=r2_score(y_test,ypred)
print("R2 Score:",sr2)
```

```python
mae=mean_absolute_error(y_test,ypred)
print("mean_absolute_error:",mae)
```

```python
mse=mean_squared_error(y_test,ypred)
print("mean_absolute_error:",mse)
```

```python
from sklearn.tree import DecisionTreeRegressor
x=df[["lotsize","bathrms"]]
y=df.price
```

```python
r2 = r2_score(y_test, y_pred)
print("r2 score:",r2)
mae = mean_absolute_error(y_test, y_pred)
print("mean_absolute_error:",mae)
mse = mean_squared_error(y_test, y_pred)
print("mean_squared_error",mse)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
x = df[["lotsize", "bathrms"]]
y = df["price"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
rf=RandomForestRegressor(n_estimators=100,random_state=42,max_depth=10,min_samples_split=5)
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred_rf=rf.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred_rf)
print("mean_squared_error",mse)
mae = mean_absolute_error(y_test, y_pred_rf)
print("mean_absolute_error",mae)
r2 = r2_score(y_test, y_pred_rf)
print("r2_score",r2)
```

```python
feature_importances = rf.feature_importances_
plt.bar(["lotsize", "bathrms"], feature_importances)
plt.title("Feature Importance")
plt.ylabel("Importance")
plt.show()
```
